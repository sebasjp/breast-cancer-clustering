```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
df = pd.read_csv("../data/sisben_2019.csv", sep=";")
```

```python
df.columns = df.columns.str.lower()
```

```python
print(df.shape)
df.head()
```

```python
cols_except = ["pared", "piso",
             "estrato", "elimbasura",
             "sanitar", "agua",
             "cocinan", "alumbra",
             "estcivil", "discapa",
             "carnet", "tipoesta",
             "nivel", "activi"]
for x in df.columns:
    x_counts = df[x].value_counts(normalize=True, dropna=False)
    if (x_counts.shape[0] <= 5) | (x in cols_except): 
        print("="*20, x, "="*20)
        print(x_counts)
    #else:
        #print(x_counts.index)
```

```python
df["tiene_paredes_ladrillo"] = (df["pared"] == 1).astype(int)
df["tiene_paredes_tapia"] = (df["pared"] == 2).astype(int)
df["tiene_paredes_tabla"] = (df["pared"] == 5).astype(int)
df["tiene_piso_cemento"] = (df["piso"] == 3).astype(int)
df["tiene_piso_baldosa"] = (df["piso"] == 2).astype(int)
df["tiene_piso_madera_tierra"] = (df["piso"].isin([4, 5])).astype(int)
df["es_estrato_1y2"] = (df["estrato"].isin([1, 2])).astype(int)
df["tiene_inodoro_alcantarillado"] = (df["sanitar"] == 1).astype(int)
df["tiene_letrina"] = (df["sanitar"].isin([3, 4])).astype(int)
df["tiene_agua_acueducto"] = (df["agua"] == 1).astype(int)
df["cocinan_electricidad"] = (df["cocinan"] == 1).astype(int)
df["cocinan_gas_natural"] = (df["cocinan"] == 2).astype(int)
df["cocinan_gas_propano"] = (df["cocinan"] == 3).astype(int)
df["tiene_eps_subsidiada"] = (df["carnet"] == 4).astype(int)
df["tiene_eps_contributiva"] = (df["carnet"] == 3).astype(int)
df["tiene_eps_ninguna"] = (df["carnet"] == 0).astype(int)
df["tiene_nivel_educativo_ninguna"] = (df["nivel"] == 0).astype(int)
df["tiene_nivel_educativo_primaria"] = (df["nivel"] == 1).astype(int)
df["tiene_nivel_educativo_secundaria"] = (df["nivel"] == 2).astype(int)
df["es_zona_urbana"] = (df["nomvereda"] == "BARRIO ZONA URBANA").astype(int)
df["tiene_servicio_telefono"] = (df["telefono"] == 1).astype(int)
df["es_vivi_casa_apto"] = (df["vivienda"] == 1).astype(int)
df["tiene_energia"] = (df["energia"] == 1).astype(int)
df["tiene_alcanta"] = (df["alcanta"] == 1).astype(int)
df["tiene_gas"] = (df["gas"] == 1).astype(int)
df["tiene_servicio_basura"] = (df["basura"] == 1).astype(int)
df["tiene_acueducto"] = (df["acueduc"] == 1).astype(int)
df["es_vivi_arriendo"] = (df["teneviv"] == 1).astype(int)
df["es_vivi_prop_pagando"] = (df["teneviv"] == 2).astype(int)
df["es_vivi_prop"] = (df["teneviv"] == 3).astype(int)
df["es_vivi_otro"] = (~df["teneviv"].isin([1, 2, 3])).astype(int)
df["tiene_sanitario_dentro_vivi"] = (df["usanitar"] == 1).astype(int)
df["tiene_sanitario_exclusivo"] = (df["usosani"] == 1).astype(int)
df["tiene_ducha"] = (df["ducha"] == 1).astype(int)
df["tiene_agua_7dias"] = (df["llega"] == 1).astype(int)
df["tiene_agua_24h"] = (df["suminis"] == 1).astype(int)
df["tiene_espacio_cocinar"] = (df["preparan"] == 1).astype(int)
df["tiene_cocina_exclusiva_vivi"] = (df["cocina"] == 1).astype(int)
df["tiene_telefono_exclusivo_vivi"] = (df["usotele"] == 1).astype(int)
df["tiene_nevera"] = (df["nevera"] == 1).astype(int)
df["tiene_lavadora"] = (df["lavadora"] == 1).astype(int)
df["tiene_televisor"] = (df["tvcolor"] == 1).astype(int)
df["tiene_televisor_parabolica"] = (df["tvcable"] == 1).astype(int)
df["tiene_calentador"] = (df["calenta"] == 1).astype(int)
df["tiene_horno"] = (df["horno"] == 1).astype(int)
df["tiene_aire"] = (df["aire"] == 1).astype(int)
df["tiene_computador"] = (df["computador"] == 1).astype(int)
df["tiene_equipo"] = (df["equipo"] == 1).astype(int)
df["tiene_moto"] = (df["moto"] == 1).astype(int)
df["tiene_tractor"] = (df["tractor"] == 1).astype(int)
df["tiene_auto"] = (df["auto1"] == 1).astype(int)
df["tiene_bieraices"] = (df["bieraices"] == 1).astype(int)
df["sexo_masculino"] = (df["sexo"] == 1).astype(int)
df["embarazada_o_con_hijos"] = (df["embaraza"] == 1).astype(int)
df["asiste_centro_educativo"] = (df["asiste"] == 1).astype(int)
df["tiene_ingresos"] = (df["percibe"] == 1).astype(int)
df["es_extranjero"] = (df["extranjero"] == 1).astype(int)
df["paga_arriendo"] = (df["pagapor"] == 1).astype(int)
df["trabaja_servicios_domestico"] = (df["serdomes"] == 1).astype(int)
```

```python
cols_drop = ["zona", "nomvereda", "telefono", "vivienda",
             "energia", "alcanta", "gas",
             "basura", "acueduc", "teneviv",
             "usanitar", "usosani", "ducha",
             "llega", "suminis", "preparan",
             "cocina", "usotele", "nevera",
             "lavadora", "tvcolor", "tvcable",
             "calenta", "horno", "aire",
             "computador", "equipo", "moto",
             "tractor", "auto1", "bieraices",
             "sexo", "embaraza", "asiste",
             "percibe", "extranjero", "pagapor",
             "serdomes",
             "comuna", "barrio", "nombarrio",
             "vereda", "fecha", "hogar",
             "orden", "tipodoc", "parentes",
             "conyuvive", "hijosde", "fechanto", 
             "grado", "depto", "munic",
             "elimbasura", "alumbra", "estcivil",
             "discapa", "tipoesta", "activi",
             "pared", "piso", "estrato", "sanitar",
             "agua", "cocinan", "carnet", "nivel"
            ]

df = df.drop(columns=cols_drop)
```

```python
df_unico = df.groupby(["formulario"]).max()
df_ingresos = df.groupby(["formulario"]).agg({"ingresos": "sum"})

df_unico = df_unico.drop(columns=["ingresos"])

print(df_unico.shape)
print(df_ingresos.shape)
df_unico = df_unico.join(df_ingresos)
```
